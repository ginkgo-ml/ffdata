```
import pandas as pd 

#Enter desired years of data
YEARS = [2019,2020,2021]


####Get Plays####

data_plays = pd.DataFrame()
for i in YEARS:  
    i_data = pd.read_csv('https://github.com/nflverse/nflfastR-data/blob/master/data/' \
                         'play_by_play_' + str(i) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)
    data_plays = data_plays.append(i_data, sort=True)
data_plays.reset_index(drop=True, inplace=True)
data_plays.to_csv('plays.csv')


####Get Yearly Player Stats####

data_player = pd.DataFrame()
for i in YEARS:  
    i_data = pd.read_csv('https://github.com/nflverse/nflfastR-data/blob/master/data/' \
                         'player_stats/player_stats_' + str(i) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)
    data_player = data_player.append(i_data, sort=True)
data_player.reset_index(drop=True, inplace=True)
data_player.to_csv('player_data.csv')


####Get Rosters####

data_roster = pd.DataFrame() 
i_data = pd.read_csv('https://github.com/nflverse/nflfastR-roster/blob/master/data/nflfastR-roster.csv.gz?raw=True', \
                    compression='gzip', low_memory=False)
data_roster = data_roster.append(i_data, sort=True)
data_roster.reset_index(drop=True, inplace=True)
data_roster.to_csv('rosters.csv')


####Get Injuries####

data_injuries = pd.DataFrame() 
i_data = pd.read_csv('https://github.com/nflverse/nflfastR-roster/blob/master/data/nflfastR-injuries.csv.gz?raw=True', \
                     compression='gzip', low_memory=False)
data_injuries = data_injuries.append(i_data, sort=True)
data_injuries.reset_index(drop=True, inplace=True)
data_injuries.to_csv('injuries.csv')


####Get Depth Charts####

data_depth_charts = pd.DataFrame() 
i_data = pd.read_csv('https://github.com/nflverse/nflfastR-roster/blob/master/data/nflfastR-depth_charts.csv.gz?raw=True', \
                     compression='gzip', low_memory=False)
data_depth_charts = data_depth_charts.append(i_data, sort=True)
data_depth_charts.reset_index(drop=True, inplace=True)
data_depth_charts.to_csv('depth_charts.csv')
```
